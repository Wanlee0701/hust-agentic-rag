```python
from src.embeddings import EmbeddingModelManager, VectorDatabaseManager
import yaml
```

```python
config_path = "./config.yaml"

with open(config_path, "r", encoding="utf-8") as f:
    config = yaml.safe_load(f)

model_manager = EmbeddingModelManager(config_path=config_path)
embeddings = model_manager.get_model()
vector_db = VectorDatabaseManager(embeddings=embeddings, config_path=config_path)

retrieval_config = config.get("retrieval", {})
k = retrieval_config.get("top_k", 3)
```

```python
query = "Tiêu chí đạt học bổng loại A là gì?"
```

```python
results = vector_db.search_similar(query, k=k, score_threshold=0)
```

```python
if not results:
    print("Khong tim thay ket qua.")
else:
    for i, (doc, score) in enumerate(results, 1):
        source = doc.metadata.get("source_file", "N/A")
        doc_type = doc.metadata.get("doc_type", "N/A")
        article_title = doc.metadata.get("article_title", "N/A")
        print(f"Result {i} | distance: {score:.4f}")
        print(f"source: {source} | type: {doc_type}")
        print(f"article_title: {article_title} \n")
        print(doc.page_content)
        print("-" * 80)
```
