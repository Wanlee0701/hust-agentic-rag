```python
import chromadb
from langchain_chroma import Chroma
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/distiluse-base-multilingual-cased-v2")
persist_directory = "./data/chroma"
vector_db = Chroma(
    persist_directory=persist_directory,
    embedding_function=embeddings,
    collection_name="student_regulations"
)
```

```python
all_data = vector_db.get(limit=5)
for i in range(len(all_data["documents"])):
    content = all_data["documents"][i]
    metadata = all_data["metadatas"][i]
    print(f"Chunk {i+1}:")
    print(f"Metadata: {metadata}")
    print(f"Content: {content}...")  # In ra 200 ký tự đầu tiên của chunk
    print("-" * 50)
```

```python
query = "Học bổng Trần Đại Nghĩa ?"
results = vector_db.similarity_search_with_relevance_scores(query, k=3)

print(f"Query: {query}")
print("=" * 80)

for i, (doc, score) in enumerate(results, 1):
    print(f"Result {i}")
    print(f"Relevance score: {score:.4f}")
    print(f"Content length: {len(doc.page_content)}")
    print(f"Metadata: {doc.metadata}")
    print("Content:")
    print(doc.page_content)
    print("-" * 80)
```

```python
all_data = vector_db.get(include=["documents", "metadatas"])
source = {meta["source_file"] for meta in all_data["metadatas"]}
for i, source in enumerate(source, 1):
    print(f"{i}. {source}")
```
